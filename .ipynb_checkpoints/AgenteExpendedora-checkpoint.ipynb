```python
'''
Aplicación: Máquina expendedora de bebidas

-- INSTRUCCIONES --

Disponible 3 tipos de bebidas al gusto: agua, soda y jugo. 
Ingrese cualquiera de los tres nombres mencionados anteriormente para empezar desde ahí, caso contrario puede escribir 'nada'
#Nota: UUse minusculas para los ingresos

'''

#Se inicializan las variables: costo, usado para almacenar la suma de costos.
#Stock, usado para comprobar si existe stock de alguna bebida
#Pagado, comprueba el estado de la máquina y si el usuario ha realizado el pago en la máquina

costo = 0
stock = '0'
pagado = '0'

'''
Acción: Bucle de repetición

Realiza la misma función que una sentencia do while, el objetivo es mantener la máquina activa para los 3 escenarios
'''

while True:
    #El usuario escoge que bebida desea o si desea nada.
    
    opcionBebida = input('Ingrese que bebida desea: ').lower()
    
   
    
    
    #Escenario cuando el usuario escogió Agua
    if(opcionBebida == 'agua'):
        agua = '1'
        if(agua == '1'):
            stock = '1'
            #Selecciona el tipo de pago, para ello el usuario ingresa: moneda o tarjeta escrito literalmente
            opcionPago = input('Como desea hacer el pago (moneda o tarjeta): ').lower()
            #Se comprueba el tipo de pago que ha realizado
            #Esta sección se ejecuta en caso que se haya realizado el pago con tarjeta
            if(opcionPago == 'moneda'):
                costo += 1
                print('1 crédito')
                pagado = '1'
                if(stock == '1' and pagado == '1'):
                    print('Se ha entregado la botella de agua')
                    costo += 1
            
            #Esta sección se ejecuta en caso que se haya realizado el pago con tarjeta
            elif(opcionPago == 'tarjeta'):
                costo += 1
                pagado = '1'
                if(stock == '1' and pagado == '1'):
                    print('Se ha entregado la botella de agua')
                    costo += 1
            else:
                print('No se ha realizado ningún pago')
                pagado = '0'
                
            
        else:
            print('La máquina expendedora no posee botellas de agua')
            
    ''' 
    ############ FINALIZA EL ESCENARIO AGUA ############
    '''
    
    if(opcionBebida == 'soda' or opcionBebida == 'gaseosa'):
        soda = '1'
        if(soda == '1'):
            stock = '1'
            #Selecciona el tipo de pago, para ello el usuario ingresa: moneda o tarjeta escrito literalmente
            opcionPago = input('Como desea hacer el pago (moneda o tarjeta): ').lower()
            #Se comprueba el tipo de pago que ha realizado
            #Esta sección se ejecuta en caso que se haya realizado el pago con tarjeta
            if(opcionPago == 'moneda'):
                costo += 1
                print('1 crédito')
                pagado = '1'
                if(stock == '1' and pagado == '1'):
                    print('Se ha entregado la lata de soda')
                    costo += 1
            
            #Esta sección se ejecuta en caso que se haya realizado el pago con tarjeta
            elif(opcionPago == 'tarjeta'):
                costo += 1
                pagado = '1'
                if(stock == '1' and pagado == '1'):
                    print('Se ha entregado la lata de soda')
                    costo += 1
            else:
                print('No se ha realizado ningún pago')
                pagado = '0'
            
        else:
            print('La máquina expendedora no posee latas de soda')
            
             
    ''' 
    ############ FINALIZA EL ESCENARIO SODA ############
    '''
    
        
    if(opcionBebida == 'jugo'):
        jugo = '1'
        if(jugo == '1'):
            stock = '1'
            #Selecciona el tipo de pago, para ello el usuario ingresa: moneda o tarjeta escrito literalmente
            opcionPago = input('Como desea hacer el pago (moneda o tarjeta): ').lower()
            #Se comprueba el tipo de pago que ha realizado
            #Esta sección se ejecuta en caso que se haya realizado el pago con tarjeta
            if(opcionPago == 'moneda'):
                costo += 1
                print('1 crédito')
                pagado = '1'
                if(stock == '1' and pagado == '1'):
                    print('Se ha entregado la botella de jugo')
                    costo += 1
            
            #Esta sección se ejecuta en caso que se haya realizado el pago con tarjeta
            elif(opcionPago == 'tarjeta'):
                costo += 1
                pagado = '1'
                if(stock == '1' and pagado == '1'):
                    print('Se ha entregado la botella de jugo')
                    costo += 1
            else:
                print('No se ha realizado ningún pago')
                pagado = '0'
            
        else:
            print('La máquina expendedora no posee botellas de jugo')
            
    ''' 
    ############ FINALIZA EL ESCENARIO JUGO ############
    '''
    
    if(opcionBebida == 'nada'):
        break

print('El costo total de la aplicación es de: ',costo)
```
